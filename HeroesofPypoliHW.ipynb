```python
#dependencies
import pandas as pd
import numpy as np
```

```python
#storing file path as variable
purchase_data = "Resources/purchase_data.csv"
```

```python
#reading the purchase data cvs file
purchase_data_df = pd.read_csv(purchase_data)
purchase_data_df.head()
```

```python
# Player Count -to count the number of players
Count_players_df = purchase_data_df.groupby("SN")["SN"].nunique()
Count_players_df.count()
```

```python
#Purchasing Analysis (Total) to display statistics of of purchase data
purchase_data_df.describe()
```

```python
#display number of items
items_df = purchase_data_df.groupby("Item Name")["Item Name"].nunique()
items_df.count()
```

```python
#Data manipulation to display the average purchase price out of the "purchase_data_df" data frame
UniqueItems = len(purchase_data_df["Item Name"].unique())
AveragePurchase = purchase_data_df["Price"].mean()
NumberPurchase = len(purchase_data_df["Item Name"])
Revenue = purchase_data_df["Price"].sum()

# New Data Frame to format column data
Purchase_Analysis_Total = pd.DataFrame({"Number of Unique Items": [UniqueItems],
                                           "Average Price": [AveragePurchase],
                                           "Number of Purchases": [NumberPurchase],
                                           "Total Revenue": [Revenue]})

# Formating new data frame
Purchase_Analysis_Total["Average Price"] = Purchase_Analysis_Total["Average Price"].map("${:.2f}".format)
Purchase_Analysis_Total["Total Revenue"] = Purchase_Analysis_Total["Total Revenue"].map("${:.2f}".format)
Purchase_Analysis_Total = Purchase_Analysis_Total[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

Purchase_Analysis_Total
```

```python
# Data manipulation to display gender demographics/percentage of players -
#Percentage and Count of Male Players Percentage and Count of Female Players Percentage and Count of Other / Non-Disclosed
duplicates = purchase_data_df.drop_duplicates(subset='SN', keep="last")
TotalGender = duplicates["Gender"].count()
MaleGender = duplicates["Gender"].value_counts()['Male']
FemaleGender = duplicates["Gender"].value_counts()['Female']
NonDiscGender = TotalGender - MaleGender - FemaleGender

MalePercentage = (MaleGender / TotalGender) * 100
FemalePercentage = (FemaleGender / TotalGender) * 100
NonDiscPercentage = (NonDiscGender / TotalGender) * 100

# Creating display of new DataFrame
Gender_Demographics = pd.DataFrame({"": ['Male', 'Female', 'Other/Non-Disclosed'],
                            "Percentage of Players": [MalePercentage, FemalePercentage, NonDiscPercentage],
                            "Total Count": [MaleGender, FemaleGender, NonDiscGender]})

#New DataFrame formatting
Gender_Demographics["Percentage of Players"] = Gender_Demographics["Percentage of Players"].map("{:.2f}%".format)
Gender_Demographics = Gender_Demographics.set_index('')
Gender_Demographics
```

```python
#  Purchasing Analysis (Gender) -Group by Gender
grouping_df = purchase_data_df.groupby(["Gender"])

#Obtaining data by manipulation
purchaseCount = grouping_df["SN"].count()
purchasePrice = grouping_df["Price"].mean()
purchaseValue = grouping_df["Price"].sum()

#Cleaning data by deleting all duplicates thus print results
duplicates = purchase_data_df.drop_duplicates(subset='SN', keep="last")
grouped_dup = duplicates.groupby(["Gender"])

# displaying cleaned data
purchaseCleanData = (grouping_df["Price"].sum() / grouped_dup["SN"].count())

#New DataFrame to extract data
Purchase_Analisis_Gen = pd.DataFrame({"Purchase Count": purchaseCount,
                              "Average Purchase Price": purchasePrice,
                              "Total Purchase Value": purchaseValue,
                              "Avg Total Purchase per Person": purchaseCleanData})

# DataFrame formatting
Purchase_Analisis_Gen["Average Purchase Price"] = Purchase_Analisis_Gen["Average Purchase Price"].map("${:.2f}".format)
Purchase_Analisis_Gen["Total Purchase Value"] = Purchase_Analisis_Gen["Total Purchase Value"].map("${:.2f}".format)
Purchase_Analisis_Gen["Avg Total Purchase per Person"] = Purchase_Analisis_Gen["Avg Total Purchase per Person"].map("${:.2f}".format)
Purchase_Analisis_Gen = Purchase_Analisis_Gen[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
Purchase_Analisis_Gen
```
